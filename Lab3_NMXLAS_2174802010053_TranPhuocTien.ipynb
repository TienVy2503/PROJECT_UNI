1. Viết chương trình biến đổi ảnh

1.1 Chọn đối tượng trong ảnh

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file
data = iio.imread('fruit.jpg')

# Cắt hình ảnh
bmg = data[800:1200, 570:980]

# In kích thước của hình ảnh
print(data.shape)

# Lưu hình ảnh đã cắt
iio.imsave('orange.jpg', bmg)

# Hiện thị hình ảnh đã cắt
plt.imshow(bmg)
plt.show()
```

1.2 Tịnh tiến đơn

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('fruit.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# Dịch hình ảnh
bdata = nd.shift(data, (100, 25))

# Hiện thị hình ảnh đã dịch
plt.imshow(bdata, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1.3 Thay đổi kích thước ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file
data = iio.imread('fruit.jpg')

# In kích thước hình ảnh ban đầu
print(data.shape)

# Dịch hình ảnh
bdata = nd.zoom(data, 2)  # Tăng kích thước gấp đôi

# In kích thước hình ảnh đã thay đổi
print(bdata.shape)

# Tạo 2 biến hình ảnh khác với kích thước khác nhau
data2 = nd.zoom(data, (2, 0.5, 1))  # Thay đổi chiều cao và giữ chiều rộng
data3 = nd.zoom(data, (0.5, 0.9, 1))  # Giảm chiều cao và thay đổi chiều rộng

# Hiện thị hình ảnh đã thay đổi kích thước
plt.imshow(data3)
plt.axis('off')  # Tắt trục
plt.show()
```

1.4 Xoay ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file
data = iio.imread('fruit.jpg')

# In kích thước hình ảnh
print(data.shape)

# Xoay hình ảnh 20 độ
d1 = nd.rotate(data, 20)

# Hiện thị hình ảnh đã xoay
plt.imshow(d1)
plt.axis('off')  # Tắt trục
plt.show()

# Xoay hình ảnh 20 độ mà không thay đổi kích thước
d2 = nd.rotate(data, 20, reshape=False)

# Hiện thị hình ảnh đã xoay mà không thay đổi kích thước
plt.imshow(d2)
plt.axis('off')  # Tắt trục
plt.show()
```

1.5 Dilation or Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# In kích thước hình ảnh
print(data.shape)

# Thực hiện phẫu thuật nhị phân (binary dilation) lần đầu
d1 = nd.binary_dilation(data)

# Hiện thị hình ảnh đã phẫu thuật
plt.imshow(d1, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()

# Thực hiện phẫu thuật nhị phân (binary dilation) với 3 lần lặp
d2 = nd.binary_dilation(data, iterations=3)

# Hiện thị hình ảnh đã phẫu thuật
plt.imshow(d2, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1.6 Coordinate Mapping

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# In kích thước hình ảnh
print(data.shape)

# Lấy chiều cao và chiều rộng của hình ảnh
V, H = data.shape

# Tạo chỉ số cho hình ảnh
M = np.indices((V, H))

# Thiết lập giá trị cho d
d = 5
q = 2 * d * np.random.rand(*M.shape) - d

# Chuyển đổi chỉ số thành kiểu integer
mp = M + q.astype(int)

# Áp dụng map_coordinates để lấy giá trị pixel từ hình ảnh
dl = nd.map_coordinates(data, mp)

# Hiện thị hình ảnh đã xử lý
plt.imshow(dl, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1.7 Biến đổi chung

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Định nghĩa hàm GeoFun
def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.sin(outcoord[0] / 10.0) + outcoord[1]
    return a, b

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# Áp dụng geometric transformation
dl = nd.geometric_transform(data, GeoFun)

# Hiện thị hình ảnh đã xử lý
plt.imshow(dl, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1. Viết chương trình chọn quả kiwi từ ảnh colorful-ripe-tropical-fruits.jpg trong thư mục exercise. Tịnh tiến quả kiwi sang phải 30 pixels.

```python
import matplotlib.pyplot as plt
import imageio.v2 as iio
import numpy as np

# Đọc hình ảnh từ thư mục exercise
image = iio.imread('exercise/colorful-ripe-tropical-fruits.jpg')

# Tạo mặt nạ cho vùng màu xanh lá cây (ví dụ: kiwi)
mask = (image[:, :, 0] < 100) & (image[:, :, 1] > 100) & (image[:, :, 2] < 100)

# Áp dụng mặt nạ để lấy quả kiwi
kiwi = np.zeros_like(image)
kiwi[mask] = image[mask]

# Tịnh tiến quả kiwi sang phải 30 pixels
shifted_kiwi = np.zeros_like(image)
shifted_kiwi[:, 30:] = kiwi[:, :-30]

# Hiển thị hình ảnh
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(kiwi)
plt.axis('off')
plt.title('Quả Kiwi')

plt.subplot(1, 2, 2)
plt.imshow(shifted_kiwi)
plt.axis('off')
plt.title('Quả Kiwi đã tịnh tiến')

plt.show()
```

5. Viết chương trình tạo menu
- Tịnh tiến
- Xoay
- Phóng to
- Thu nhỏ
- Coordinate Map

Khi chọn phím T, X, P, H, C thì hỏi muốn thực hiện trên hình nào từ 3 hình trong thư mục exercise. Người dùng chọn hình nào thì thực hiện phép biến đổi trên hình đó.

```python
import matplotlib.pyplot as plt
import imageio.v2 as iio
import numpy as np
import cv2

# Danh sách hình ảnh có sẵn
image_files = {
    '1': 'exercise/colorful-ripe-tropical-fruits.jpg',
    '2': 'exercise/ha-long-bay-in-vietnam.jpg',
    '3': 'exercise/pagoda.jpg'
}

def choose_image():
    print("Chọn hình ảnh:")
    print("1. Fruits")
    print("2. Ha Long Bay")
    print("3. Pagoda")
    choice = input("Nhập số 1-3: ")
    if choice in image_files:
        return iio.imread(image_files[choice])
    else:
        print("Lựa chọn không hợp lệ.")
        return None

def translate_image(img, tx=50, ty=50):
    rows, cols = img.shape[:2]
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (cols, rows))

def rotate_image(img, angle=45):
    rows, cols = img.shape[:2]
    center = (cols // 2, rows // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1)
    return cv2.warpAffine(img, M, (cols, rows))

def scale_image(img, scale=1.5):
    return cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

def shrink_image(img, scale=0.5):
    return cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_AREA)

def coordinate_map(img):
    rows, cols = img.shape[:2]
    map_x = np.zeros((rows, cols), np.float32)
    map_y = np.zeros((rows, cols), np.float32)

    for i in range(rows):
        for j in range(cols):
            map_x[i, j] = cols - j - 1  # lật ngang
            map_y[i, j] = rows - i - 1  # lật dọc

    return cv2.remap(img, map_x, map_y, cv2.INTER_LINEAR)

def show_image(title, img):
    plt.imshow(img)
    plt.axis('off')
    plt.title(title)
    plt.show()

# Menu điều khiển
print("MENU:")
print("T - Tịnh tiến")
print("X - Xoay")
print("P - Phóng to")
print("H - Thu nhỏ")
print("C - Coordinate Map")
action = input("Chọn thao tác (T/X/P/H/C): ").upper()

# Thực hiện thao tác
image = choose_image()
if image is not None:
    if action == 'T':
        result = translate_image(image)
        show_image("Tịnh Tiến", result)
    elif action == 'X':
        result = rotate_image(image)
        show_image("Xoay", result)
    elif action == 'P':
        result = scale_image(image)
        show_image("Phóng To", result)
    elif action == 'H':
        result = shrink_image(image)
        show_image("Thu Nhỏ", result)
    elif action == 'C':
        result = coordinate_map(image)
        show_image("Coordinate Map", result)
    else:
        print("Lựa chọn thao tác không hợp lệ.")
```
